So I encountered a new problem with this one. The thing is, median, when stuck between two values in an even-numbered array, will resort to mean of the two central values. And that may give us a bunch of 'median' values in place of the missing ones, which normally couldn't occur by design of that column's values. In this case, we would get values like 4.5 or 7 or 9 cylinders, which weren't in the column before (probably because engines with 4.5/7/9 cylinders do not exist!). 

I decided to combine the median and mode functions to get the values that we need. Going through the rows with a loop and checking if the resulting median value strays from the list of unique values in the column will allow us to rely on mode() in that case, which will simply select the most popular value in the group (our next best option where median doesn't work).

I'm filling in the values in a new column based on 'cylinders', just so we can compare them on a single histogram against each other after we have the results.

```python
models = data['model'].unique()
cylinders = data['cylinders'].unique()
data['newC'] = data['cylinders']

for model in models:
    median_c = data.loc[data['model'] == model]['cylinders'].median()
    if median_c in cylinders:
        data.loc[data['model'] == model, 'newC'] = data['newC'].fillna(median_c)
    else:
        mode_c = data.loc[data['model'] == model]['cylinders'].mode()
        data.loc[data['model'] == model, 'newC'] = data['newC'].fillna(mode_c)
```

Looks good. The resulting distribution is line with the original data, so it seems like it worked well, and we don't have any outlandish values like 4.5 cylinders. We can now fill in the original column using the same method and get rid of the extra column.

```python
models = data['model'].unique()
cylinders = data['cylinders'].unique()

for model in models:
    median_c = data.loc[data['model'] == model]['cylinders'].median()
    if median_c in cylinders:
        data.loc[data['model'] == model, 'cylinders'] = data['cylinders'].fillna(median_c)
    else:
        mode_c = data.loc[data['model'] == model]['cylinders'].mode()
        data.loc[data['model'] == model, 'cylinders'] = data['cylinders'].fillna(mode_c)
```

Everything seems to be in order! Now let's do miles_per_year. For this one we'll use the apply() function so as not to divide those cars who's age in years is zero by their mileage in one swoop with the older cars. This way we can filter out the 0-aged cars to simply assign whatever mileage they have in total to their miles_per_year. In this case, we must pass the entire rows to the function so that we use values from different columns.

```python
def make_miles_per_year(row):
    age = row['age']
    total_miles = row['mileage']
    
    if age == 0:
        return total_miles
    else:
        return int(total_miles / age)
    
data['miles_per_year'] = data.apply(make_miles_per_year, axis = 1)
```

### Conclusion

From what we have seen, age and mileage appear to be the factors most correlated with price of a car in our dataset. It's helpful to have all types of graphs that are appropriate for the types of data involved, as they can help us see the relationship in a more rounded way. For instance, we can definitely see the relationship by how curved the line is if we plot median values, as opposed to a proper scatterplot.

The relationship between age and mileage and price is exactly what one would expect: the newer the car, the better its value on the market. And we could easily expect in 9 cases out of 10 for mileage to be closely correlated with age, for obvious reasons.

It's a little more puzzling to see that the corr() coefficient for condition and price appears to be two times weaker than for mileage or age, even though we can see the relationship more or less clearly on the graph. Remember that our condition scale goes from 0 to 5 where 0 is salvage and 5 is (supposedly) new. For each of the 4 categories in between there appears to be a clear progression of these data points intersecting more often with higher values for price. I think there are two reasons for that (for the weaker correlation). Firstly, condition being a categorical value, we had to use boxplot, which is helpful, but a little deceiving in that it is easy to spot the progression, which is there, but harder to remind yourself that the boxes' ranges actually still overlap for (what looks like) the most part. After all, there are only so many conditions (6, or 4 with enough data to estimate correlation), and they inevitably capture a wider range of prices. The second reason for that is that condition, as it is originally collected and reflected in our dataset, is not as precise and reliable a value as mileage and age of the car, but is decided upon by the seller and is wildly subjective compared to the reading of the car's odometer. Perhaps if there was some kind of a precise condition index, computed based on age, mileage, number and severity of scratches, repairs, recorded accidents and so on, it would actually correlate with the price much more strongly.


Transmission is a category for which we have disproportionate amount of data for the two types (automatic and manual), and too little data for mixed transmission, to see a difference in relationship. Based on the data we have, however, there is a difference for sedans and SUVs: it's quite self-explanatory for sedans that manual transmission cars are in most cases the more outdated models, and since they are not as well suited for the most drivers' purposes, they are less in demand. Therefore their median price is a couple thousand lower. With SUVs the opposite appears to be the case: they are at least two thousand or so more expensive, judging by the median value (however it is clear from the graph that most of their price values are, in fact, in the same range). If a corr() value could be computed for transmission and price, I wouldn't expect it to be very high, since the two values (if we don't count the mixed type), cover almost every other possible combination of factors. But I suppose it makes sense that manual transmission is more suited for some people's purposes and demands from a car. Since it implies more precise control over transmission, it may well be something that could be valuable to SUV owners more often, so for that type of car it's more a valuable feature, than a useless and cumbersome transmission mechanism, as it is likely perceived by unassuming sedan-owning city drivers.

For colors it's bit more tricky, since we can definitely guess that color is a factor that likely doesn't determine a car's price directly, but can rather be associated with a set of other features that do that. For instance, red or yellow could in some cases be disproportionately more popular with some kind of expensive sportcar-like vehicles.
